```python
def create_address(first_name, last_name, street, number, city, zipcode):
    deliver_word = "deliver" if first_name == 'John' else "Deliver"
    return f"Hand- {deliver_word} to: {first_name} {last_name}\n {street} Apt. {number}\n{zipcode} {city}"
```
