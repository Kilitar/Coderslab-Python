```python
import psycopg2

connection = psycopg2.connect(
    host='localhost',
    user='postgres',  # correct this if necessary
    password='postgres',  # correct this if necessary
    dbname='classicmodels',  # correct this if necessary
)

cursor = connection.cursor()

cursor.execute('SELECT city, addressLine1, addressLine2, state, country FROM offices ORDER BY city')

for row in cursor:
    print(f'Office city {row[0]}:')
    print(row[1])

    if row[2] is not None:  # we only display address 2nd line if the office has it
        print(row[2])
    
    if row[3] is None:  # if there is no state/region in the address display only country
        print(row[4])
    else:
        print(f'{row[3]}, {row[4]}')
    
    print('')
```
