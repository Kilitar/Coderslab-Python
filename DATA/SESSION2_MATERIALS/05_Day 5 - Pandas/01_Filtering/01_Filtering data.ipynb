# Filtering data
While generating the `product_prices_renamed.csv` file the following processing errors occurred and they need to be found:
1. In the **date** column, data from 1888 appeared - '1888-0',
2. Check the **date** column for similar errors but with future values.
3. In the **value** column, too high value was introduced – find it and locate the row where it is (use `query`),
4. There was a spelling error in the **product_types** column for one of the products. Find it and the corresponding rows. How many such rows are there?

You do not need to assign the results to the variable: it is enough if you display them.

> Based on the solution of this task, we will later correct all errors in the data.

Hints:

2. Use `loc` or `query` with the condition `date > '2020-1'`.

Subsection 3:
1. There is only one such value.
1. Do the following:
a) use `describe()`, to view percentiles,
b) use `loc` or `query` to find erroneous entries,

Subsection 4:

You can do it in the following way:
a) use `unique()` method to find all available values,
b) use `loc` or `query` to find erroneous entries,
c) The number of rows can be checked with the `shape` method.

```python
import pandas as pd

# 0. Načtení dat
df = pd.read_csv("../01_Data/product_prices_renamed.csv", sep=";")
```

## Varianta loc

```python
# 1. In the date column, data from 1888 appeared - '1888-0'
print("=== Varianta loc: 1. Chyby s rokem 1888 ===")
err_1888_loc = df.loc[df["date"] == "1888-0"]
print(f"Počet řádků: {err_1888_loc.shape[0]}")
print(err_1888_loc.head())

# 2. Check the date column for similar errors but with future values
print("\n=== Varianta loc: 2. Chyby s budoucími daty ===")
err_future_loc = df.loc[df["date"] > "2020-1"]
print(f"Počet řádků: {err_future_loc.shape[0]}")
print(err_future_loc.head())

# 3. In the value column, too high value was introduced
print("\n=== Varianta loc: 3. Extrémně vysoké hodnoty ===")
print(df["value"].describe())
err_high_loc = df.loc[df["value"] > 500]
print(f"\nPočet řádků s extrémní hodnotou: {err_high_loc.shape[0]}")
print(err_high_loc.head())

# 4. Spelling error in product_types column
print("\n=== Varianta loc: 4. Překlep v product_types ===")
err_misspelled_loc = df.loc[df["product_types"] == "fresh chichen egges - per 666pcs."]
print(f"Počet řádků s překlepem: {err_misspelled_loc.shape[0]}")
print(err_misspelled_loc.head())
```

## Varianta query

```python
# 1. In the date column, data from 1888 appeared - '1888-0'
print("=== Varianta query: 1. Chyby s rokem 1888 ===")
err_1888_query = df.query("date == '1888-0'")
print(f"Počet řádků: {err_1888_query.shape[0]}")
print(err_1888_query.head())

# 2. Check the date column for similar errors but with future values
print("\n=== Varianta query: 2. Chyby s budoucími daty ===")
err_future_query = df.query("date > '2020-1'")
print(f"Počet řádků: {err_future_query.shape[0]}")
print(err_future_query.head())

# 3. In the value column, too high value was introduced
print("\n=== Varianta query: 3. Extrémně vysoké hodnoty ===")
err_high_query = df.query("value > 500")
print(f"Počet řádků s extrémní hodnotou: {err_high_query.shape[0]}")
print(err_high_query.head())

# 4. Spelling error in product_types column
print("\n=== Varianta query: 4. Překlep v product_types ===")
err_misspelled_query = df.query("product_types == 'fresh chichen egges - per 666pcs.'")
print(f"Počet řádků s překlepem: {err_misspelled_query.shape[0]}")
print(err_misspelled_query.head())
```
