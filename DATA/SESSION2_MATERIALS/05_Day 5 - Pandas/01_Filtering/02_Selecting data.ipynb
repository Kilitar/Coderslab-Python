## Selection of data for analysis

Answer the following questions:

1. What was the average price of a carrot and pea mix per 1 kg in Poland in 2015?
2. What was the average price of apple juice in 2016-2018 in the Masovia province?
3. What was the average price of tomato paste in Lower Silesia Province in 2003-2015? Compared with other products does it seem reasonable to you?<br>

Suggest what can be done with values equal to 0. How does this affect the results of point 3?

> All the product groups needed for this task are in the **product_types** column.

```python
import pandas as pd

# Načtení dat
df = pd.read_csv("../01_Data/product_prices_renamed.csv", sep=";")
df["year"] = df["date"].str.split("-").str[0]
```

### Q1: Average price of frozen carrot and pea mix per 1 kg in Poland in 2015

```python
# Varianta loc
q1_loc = df.loc[
    (df["product_types"] == "frozen carrot and pea mix - per 1kg") &
    (df["province"] == "POLAND") &
    (df["year"] == "2015")
]
print("Varianta loc - průměrná cena:", q1_loc["value"].mean())

# Varianta query
q1_query = df.query("product_types == 'frozen carrot and pea mix - per 1kg' and province == 'POLAND' and year == '2015'")
print("Varianta query - průměrná cena:", q1_query["value"].mean())
```

### Q2: Average price of apple juice in 2016-2018 in Masovia province

```python
# Varianta loc
q2_loc = df.loc[
    (df["product_types"] == "apple juice, boxed - per 1l") &
    (df["province"] == "MASOVIA") &
    (df["year"].isin(["2016", "2017", "2018"]))
]
print("Varianta loc - průměrná cena:", q2_loc["value"].mean())

# Varianta query
q2_query = df.query("product_types == 'apple juice, boxed - per 1l' and province == 'MASOVIA' and year in ['2016', '2017', '2018']")
print("Varianta query - průměrná cena:", q2_query["value"].mean())
```

### Q3: Average price of tomato paste in Lower Silesia in 2003-2015

```python
years_2003_2015 = [str(y) for y in range(2003, 2016)]

# Varianta loc
q3_loc = df.loc[
    (df["product_types"] == "30% tomato concentrate - per 1kg") &
    (df["province"] == "LOWER SILESIA") &
    (df["year"].isin(years_2003_2015))
]

print("Průměrná cena (včetně nul):", q3_loc["value"].mean())
print("Průměrná cena (bez nul):", q3_loc.loc[q3_loc["value"] > 0, "value"].mean())

# Zobrazení počtu nulových a nenulových hodnot
print("Počet nulových hodnot:", (q3_loc["value"] == 0).sum())
print("Počet nenulových hodnot:", (q3_loc["value"] > 0).sum())
print("Maximální hodnota v tomto subsetu:", q3_loc["value"].max())
```
