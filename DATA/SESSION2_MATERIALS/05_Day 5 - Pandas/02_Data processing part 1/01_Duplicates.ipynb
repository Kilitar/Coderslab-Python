# Duplicates
Using the **product_prices_renamed.csv** files:
1. check how many rows are duplicated,
2. using `duplicated` check which rows were doubled.
3. using `drop_duplicates`, remove all duplicates from the output dataset (we assume that a row with a duplicate is an error and needs to be removed) – write the result to a new DataFrame.

```python
import pandas as pd

# Načtení dat
df = pd.read_csv("../01_Data/product_prices_renamed.csv", sep=";")

# 1. Počet duplicitních řádků
num_duplicates = df.duplicated().sum()
print("Počet duplicitních řádků:", num_duplicates)

# 2. Které řádky jsou zdvojené
duplicated_rows = df.loc[df.duplicated(keep=False)]
print("\nZdvojené řádky (ukázka):")
print(duplicated_rows.head())

# 3. Odstranění duplicit a zápis do nového DataFrame
df_clean = df.drop_duplicates()
print("\nPůvodní počet řádků:", df.shape[0])
print("Počet řádků po odstranění duplicit:", df_clean.shape[0])
```
