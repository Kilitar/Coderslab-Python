# Highest and lowest price

Using `sort_values` and the data from the _product_prices_renamed.csv_ the file answer the questions below — use data aggregated for entire Poland (`province=='POLAND'`):

1. What product had the highest 'PLN' price in history? (remember to remove the value = 3000)?
2. What product achieved the lowest 'PLN' price in history? (remember to filter out values = 0),

Sort the data according to the following key:

1. by **product_types** column – descending
2. by **value** column – ascending

```python
import pandas as pd

# Načtení dat
df = pd.read_csv("../01_Data/product_prices_renamed.csv", sep=";")

# Filtrování pouze pro celé Polsko
df_poland = df.loc[df["province"] == "POLAND"].copy()
```

### 1. Nejvyšší cena v historii (vyjma 3000)

```python
print("=== A. Zohlednění všech produktů (i s nevyplněným product_types, tj. přes product_line) ===")
highest_all = df_poland.loc[df_poland["value"] != 3000].sort_values(by="value", ascending=False).head(1)
print(highest_all[["product_types", "product_line", "value", "date"]])

print("\n=== B. Zohlednění pouze produktů s vyplněným product_types ===")
highest_pt = df_poland.loc[(df_poland["value"] != 3000) & (df_poland["product_types"].notna())].sort_values(by="value", ascending=False).head(1)
print(highest_pt[["product_types", "value", "date"]])
```

### 2. Nejnižší nenulová cena v historii

```python
print("=== A. Zohlednění všech produktů (přes product_line, vyjma 0) ===")
lowest_all = df_poland.loc[df_poland["value"] > 0].sort_values(by="value", ascending=True).head(1)
print(lowest_all[["product_types", "product_line", "value", "date"]])

print("\n=== B. Zohlednění pouze produktů s vyplněným product_types ===")
lowest_pt = df_poland.loc[(df_poland["value"] > 0) & (df_poland["product_types"].notna())].sort_values(by="value", ascending=True).head(1)
print(lowest_pt[["product_types", "value", "date"]])
```

### 3. Seřazení dat
Klíč:
* `product_types` – sestupně (descending)
* `value` – vzestupně (ascending)

```python
sorted_df = df_poland.sort_values(by=["product_types", "value"], ascending=[False, True])
print(sorted_df[["product_types", "value", "date"]].head(10))
```
