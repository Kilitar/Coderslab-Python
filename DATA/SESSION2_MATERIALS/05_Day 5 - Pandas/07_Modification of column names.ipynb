# Modification of column names

Read the file **product_prices.csv** into a DataFrame. Modify the names of columns so that they match those in the **product_prices_renamed.csv** file. Complete two variants of the exercise:

1. Using the `rename` method,
2. Overwriting all column names using `columns`.

### Target columns mapping:
* `'Name'` -> `'province'`
* `'Goods types'` -> `'product_types'`
* `'Measurement unit'` -> `'currency'`
* `'Group ID'` -> `'product_group_id'`
* `'Product types'` -> `'product_line'`
* `'Value'` -> `'value'`
* `'Date'` -> `'date'`

```python
import pandas as pd

# Načtení kontrolního souboru pro ověření
df_expected = pd.read_csv("01_Data/product_prices_renamed.csv", sep=";")
expected_cols = list(df_expected.columns)
print("Cílové sloupce:", expected_cols)
```

## 1. Varianta: rename
Použití metody `.rename(columns=...)` na nově načteném DataFrame.

```python
# Načtení původních dat
df_rename = pd.read_csv("01_Data/product_prices.csv", sep=";")

# Slovník pro přejmenování
mapping = {
    'Name': 'province',
    'Goods types': 'product_types',
    'Measurement unit': 'currency',
    'Group ID': 'product_group_id',
    'Product types': 'product_line',
    'Value': 'value',
    'Date': 'date'
}

# Přejmenování sloupců
df_rename = df_rename.rename(columns=mapping)

# Ověření shodnosti
print("Shodují se sloupce s očekávanými?", list(df_rename.columns) == expected_cols)
print(df_rename.head(3))
```

## 2. Varianta: columns
Přímé přepsání celého seznamu názvů sloupců přes atribut `.columns`.

```python
# Načtení původních dat znovu
df_columns = pd.read_csv("01_Data/product_prices.csv", sep=";")

# Přímé přepsání názvů sloupců (musí přesně odpovídat počtu a pořadí sloupců)
df_columns.columns = [
    'province',
    'product_types',
    'currency',
    'product_group_id',
    'product_line',
    'value',
    'date'
]

# Ověření shodnosti
print("Shodují se sloupce s očekávanými?", list(df_columns.columns) == expected_cols)
print(df_columns.head(3))
```
