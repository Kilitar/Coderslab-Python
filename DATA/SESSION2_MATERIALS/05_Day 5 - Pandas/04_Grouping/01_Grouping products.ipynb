# Grouping products
Using the data from the  **product_prices_cleaned.csv** file for provinces, do the following:
1. What was the average monthly price of each commodity?
2. Which product had the highest price volatility over the years?

Use the **product** and **value** columns for analysis.
Additionally:
1. Consider whether any more assumptions are needed for the tasks.
2. Why can this task be done only now, after cleaning the data? Do you think any more operations are needed?

```python
import pandas as pd

# Načtení čistých dat
df = pd.read_csv("../01_Data/product_prices_cleaned.csv", sep=";")

# Filtrování dat POUZE pro provincie (bez agregovaného 'POLAND')
df_prov = df[df['province'] != 'POLAND']

# 1. Průměrná cena pro každý produkt
avg_price = df_prov.groupby('product')['value'].mean()
print("=== Průměrná cena pro každý produkt ===")
print(avg_price)

# 2. Nejvyšší volatilita (směrodatná odchylka)
volatility = df_prov.groupby('product')['value'].std()
highest_volatility_product = volatility.idxmax()
highest_volatility_value = volatility.max()
print("\n=== Nejvyšší volatilita ===")
print(f"Produkt: {highest_volatility_product}")
print(f"Směrodatná odchylka (volatilita): {highest_volatility_value}")
```
