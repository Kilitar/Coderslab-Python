# Removing errors
Using the solutions from the previous class and the **product_prices_renamed.csv** file, modify `loc` so that it corrects the errors present in the dataset:
1. In the **date** column, data from 1888 appeared - '1888-0', change the value to 1999-1,
2. In the **date** column, data from 2099 appeared - '2099-13', change the value to 2019-1,
3. There is a spelling error in the **product_types** column - correct it. Number of pieces should be '10pcs.'. Check whether the task was done correctly.
4. Use `loc` to convert the values given in `EUR` to `PLN` with 4.15 exchange rate.
   Instead of writing `loc` twice, first query the data for rows where **currency** = `EUR` and save it to a variable.
5. Filter from the set those rows where the price for the product is 3000.

```python
import pandas as pd

# Načtení dat a odstranění duplicit z předchozí úlohy
df = pd.read_csv("../01_Data/product_prices_renamed.csv", sep=";").drop_duplicates()

# 1. 1888-0 -> 1999-1
df.loc[df['date'] == '1888-0', 'date'] = '1999-1'

# 2. 2099-13 -> 2019-1
df.loc[df['date'] == '2099-13', 'date'] = '2019-1'

# 3. Oprava překlepu v product_types
bad_spelling = 'fresh chichen egges - per 666pcs.'
good_spelling = 'fresh chicken eggs - per 10pcs.'
df.loc[df['product_types'] == bad_spelling, 'product_types'] = good_spelling

print("Kontrola překlepu, počet špatných:", (df['product_types'] == bad_spelling).sum())
print("Kontrola překlepu, počet správných:", (df['product_types'] == good_spelling).sum())

# 4. Převod EUR na PLN (kurz 4.15)
eur_mask = df['currency'] == 'EUR'
df.loc[eur_mask, 'value'] = df.loc[eur_mask, 'value'] * 4.15
df.loc[eur_mask, 'currency'] = 'PLN'

print("\nPočet záznamů v EUR po konverzi:", (df['currency'] == 'EUR').sum())

# 5. Filtrování řádků s hodnotou 3000
df = df.loc[df['value'] != 3000]
print("Počet záznamů s cenou 3000 po filtrování:", (df['value'] == 3000).sum())
```
