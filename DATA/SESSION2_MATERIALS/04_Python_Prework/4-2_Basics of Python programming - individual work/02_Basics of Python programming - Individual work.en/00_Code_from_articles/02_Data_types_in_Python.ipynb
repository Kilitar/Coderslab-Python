#### List length

```python
club_members = ["Ike", "Mike", "Spike"]
club_members_length = len(club_members)
print(club_members_length)

text = "blah"
text_length = len(text)
print(text_length)
```

### Adding elements to lists

```python
numbers = [1, 2, 3]
print(numbers)

numbers.append(4)
print(numbers)

numbers.extend([5, 6])  # equivalent to `numbers + [5, 6]`
print(numbers)

numbers.insert(2, 42)  # insert number 42 on index 2; rest of the numbers moves one rightprint(numbers)
```
