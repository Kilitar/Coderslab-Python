# Program flow control

## Conditional statements

```python
password = input("Enter password: ")

if len(password) < 5:
    print("Password is too short!")
elif len(password) > 10:
    print("Password is too long!")
else:
    print("Perfect password!")
```

## Loops

### `for` loop

```python
data = [1, 2, 3]

for element in data:
    print(f"Element : {element}")
```

```python
heroes = [
    "Luke Skywalker",
    "Princess Leia",
    "Han Solo",
    "Chewbacca",
    "Obi-Wan Kenobi"
]

for character in heroes:
    print(character)
```

### `range`

```python
for i in range(10):
    print(i)
```

```python
for i in range(5, 10):
    print(i)
```

```python
for i in range(0, 10, 2):
    print(i)
```

```python
for i in range(10, 0, -1):
    print(i)
```

```python
result = []
for i in range(10):
    result.append(i)

print(result)
```

```python
result = list(range(10))

print(result)
```

### `while` loop

```python
i = 0
while i < 10:
    print(i)
    i += 1  # short for i = i + 1
```

```python
import random

i = 0
while i != 5:
    print(i)
    i = random.randint(1,10)
```

```python
import random

i = 0
while i < 3:
    print(i)
    if random.randint(1,10) == 5:
         print("break")
         break
    i += 1
```

```python
sum_even = 0
for i in range(10):
    if i % 2:
        print(f"i = {i} skip odd numbers!")
        continue

    print(f"i = {i}")
    sum_even += i

print(f"The sum of even numbers in range 0 - 9: {sum_even}")
```

### `for-else` statement

```python
numbers = [1, 3, 5]

for i in numbers:
    if i % 2 == 0:
        print("Even number found!")
        break

else:
    print("There are no even numbers on the list!")
```

```python
numbers = [1, 2, 5]

for i in numbers:
    if i % 2 == 0:
        print("Even number found!")
        break

else:
    print("There are no even numbers on the list!")
```
