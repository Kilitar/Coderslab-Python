# Dataset normalization

In the sheets of the **dims.xlsx** file there are dictionaries for the data from the **product_prices_cleaned.csv** file. Use `merge` to normalize the data following the steps:

1. Read the contents of the **dims.xlsx** file sheets to separate `DataFrames`.
For readability base names of frames on the names of sheets.

1. Read the data from **product_prices_cleaned.csv** file to the `df` variable.

1. Based on the **d_province** workbook, use the `id` column to add the `province_id` column to the `df` frame.

1. Based on the  **d_product** workbook, add the `product_id` column to the `df` frame.

1. From the table, extract only the columns that refer to other tables, e.g.. **product_id** and the columns **value**, **date**. Do you think this is more readable? What are potential benefits of this approach?

> We will tell you how to read many workbooks at once when we discuss `openpyxl`.

You can find more about database normalization at the [link](https://www.sqlshack.com/what-is-database-normalization-in-sql-server/).

```python
import pandas as pd

# 1. Read the contents of the dims.xlsx file sheets to separate DataFrames.
# For readability base names of frames on the names of sheets.
dims_path = '../01_Data/dims.xlsx'
d_province = pd.read_excel(dims_path, sheet_name='d_province')
d_product = pd.read_excel(dims_path, sheet_name='d_product')
d_product_group = pd.read_excel(dims_path, sheet_name='d_product_group')

print("Loaded DataFrames:")
print("d_province shape:", d_province.shape)
print("d_product shape:", d_product.shape)
print("d_product_group shape:", d_product_group.shape)

# 2. Read the data from product_prices_cleaned.csv file to the df variable.
# The cleaned dataset is located in the Day 5 data directory.
prices_path = '../../05_Day 5 - Pandas/01_Data/product_prices_cleaned.csv'
df = pd.read_csv(prices_path, sep=';')
print("\nRaw df shape:", df.shape)

# 3. Based on the d_province workbook, use the id column to add the province_id column to the df frame.
df = df.merge(d_province, on='province', how='left')

# 4. Based on the d_product workbook, add the product_id column to the df frame.
df = df.merge(d_product[['product_id', 'product']], on='product', how='left')

# 5. From the table, extract only the columns that refer to other tables, e.g.. product_id and the columns value, date.
normalized_df = df[['province_id', 'product_id', 'value', 'date']]
print("\nNormalized DataFrame head:")
print(normalized_df.head())

```

### Ad 5. Do you think this is more readable? What are potential benefits of this approach?

**Readability**:
For humans, this normalized table is less readable because it contains abstract IDs (like `province_id = 8`) instead of human-friendly names (like "SUBCARPATHIA").

**Potential Benefits**:
1. **Redundancy Reduction**: It eliminates storing repeating string data (e.g. text for province names, product names) across millions of rows, significantly saving storage space.
2. **Data Integrity & Ease of Maintenance**: If a product name or province name changes, we only need to update it once in the dimension sheet (`dims.xlsx`), and the fact table automatically reflects the update.
3. **Query Performance**: Numerical IDs are faster to index and perform joins on than text columns, improving database search and aggregation performance.
4. **Consistency**: Avoids potential typos or spelling differences (e.g. casing/whitespace inconsistencies) in text fields from causing errors during data grouping or filtering.
